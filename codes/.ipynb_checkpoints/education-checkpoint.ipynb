```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

# Source: Wikipedia

```python
url="https://en.wikipedia.org/wiki/List_of_colleges_and_universities_in_New_Jersey#Public_colleges_and_universities"
colleges_universities = pd.read_html(url)
colleges_universities
```

```python
public_coll=colleges_universities[1]
#public_coll
public_df=public_coll[["School", "Location", "Type[9]", "Control[9]"]]
public_df=public_df.rename(columns = {'Type[9]':'Type','Control[9]':'Control' }) 
public_df.head()
```

```python
private_coll=colleges_universities[2]
#private_coll
private_df=private_coll[["School", "Location", "Type[9]", "Control[9]"]]
private_df=private_df.rename(columns = {'Type[9]':'Type','Control[9]':'Control' }) 
   
private_df.head()
```

```python
profit_insti=colleges_universities[3]
profit_insti
profit_df=profit_insti[["School", "Location", "Type[9]","Control[9]"]]
profit_df=profit_df.rename(columns = {'Type[9]':'Type','Control[9]':'Control' }) 
   
profit_df.head()
```

```python
religious_coll=colleges_universities[4]
religious_coll
religious_df=religious_coll[["School", "Location", "Type[9]","Control[9]"]]
religious_df=religious_df.rename(columns = {'Type[9]':'Type','Control[9]':'Control' }) 
religious_df.head()
```

```python
theological_schools=colleges_universities[5]
theological_schools
theological_df=theological_schools[["School", "Location", "Accreditation"]]
theological_df.head()
```

```python
abbreviations=colleges_universities[6]
abbreviations
```

```python
talmudic_schools=colleges_universities[7]
talmudic_schools
talmudic_df=talmudic_schools[["School", "Location", "Accreditation"]]
talmudic_df.head()
```

```python
defunct=colleges_universities[8]
defunct.head()
```

```python
pub_pri_pro_df=pub_pri_df.append(profit_df, ignore_index=True)
pub_pri_pro_df.head()
```

```python
pub_pri_pro_rel_df=pub_pri_pro_df.append(religious_df,ignore_index=True)
pub_pri_pro_rel_df.head()
```
