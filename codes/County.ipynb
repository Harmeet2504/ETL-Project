```python
import requests
import json
import pandas as pd
import numpy as np
from config import api_key
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
#write api call for 2017 data and get response
long_url2 = "https://api.census.gov/data/2017/acs/acs5/profile?get=DP05_0001E,DP02_0037PE,DP02_0061PE,DP03_0009PE,DP03_0128PE,DP03_0063E,DP05_0004E,DP05_0018E,NAME&for=county:*&in=state:34&key="
response2 = requests.get(long_url2)
print(response2)
```

```python
#read json
response2=response2.json()
```

```python
#convert to dataframe and clean up header
census1_pd_2017 = pd.DataFrame(response2)
census1_pd_2017.columns = census1_pd_2017.iloc[0]
census1_pd_2017 = census1_pd_2017.iloc[1:]
census1_pd_2017.head()
```

```python
#rename columns
census1_pd_2017 = census1_pd_2017.rename(columns={"DP05_0001E": "2017_Population","DP02_0037PE": "2017_Birth_to_Unmarried_Women",
                                                  "DP02_0061PE": "High_School_Graduate_Percentage","DP03_0009PE":"Unemployment_Percentage",
                                                  "DP03_0128PE":"Percent_Under_the_Poverty_level","DP03_0063E":"Mean_Income",
                                                  "DP05_0004E":"Sex_Ratio_per_hundred_female","DP05_0018E":"Mean_Age"})
census1_pd_2017.head()
```

```python
census1_pd_2017.drop('state', axis=1, inplace=True)
```

```python
census1_pd_2017.drop('county', axis=1, inplace=True)
```

```python
census1_pd_2017.head()
```

```python
census1_pd_2017=census1_pd_2017[['NAME','2017_Population','2017_Birth_to_Unmarried_Women','High_School_Graduate_Percentage','Unemployment_Percentage',
                                 'Percent_Under_the_Poverty_level','Mean_Income','Sex_Ratio_per_hundred_female','Mean_Age']]
census1_pd_2017.head()
```

```python
census1_pd_2017= census1_pd_2017.rename(columns={"NAME": "county"})
census1_pd_2017.head()
```

```python
census1_pd_2017['county'] = census1_pd_2017['county'].str.rstrip('New Jersey')
census1_pd_2017.head()
```

```python
census1_pd_2017['county'] = census1_pd_2017['county'].str.rstrip('County,')
```
