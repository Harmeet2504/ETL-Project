```python
import requests
import json
import pandas as pd
import numpy as np
from config import api_key
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
#write api call for 2017 data and get response
long_url2 = "https://api.census.gov/data/2017/acs/acs5/profile?get=DP05_0001E,DP02_0037PE,DP02_0061PE,DP03_0009PE,DP03_0128PE,DP03_0063E,DP05_0004E,DP05_0018E,NAME&for=county:*&in=state:34&key="
response2 = requests.get(long_url2)
print(response2)
```

```python
#read json
response2=response2.json()
```

```python
#convert to dataframe and clean up header
census1_pd_2017 = pd.DataFrame(response2)
census1_pd_2017.columns = census1_pd_2017.iloc[0]
census1_pd_2017 = census1_pd_2017.iloc[1:]
census1_pd_2017
```

```python
#rename columns
census1_pd_2017 = census1_pd_2017.rename(columns={"DP05_0001E": "2017 Pop","DP02_0037PE": "2017 Birth to Unmarried Women",
                                                  "DP02_0061PE": "High School Graduate(%)","DP03_0009PE":"Unemployment(%)",
                                                  "DP03_0128PE":"Percent Under the Poverty lavel","DP03_0063E":"Mean Income",
                                                  "DP05_0004E":"Sex Ratio(per hundred female)","DP05_0018E":"Mean Age"})
census1_pd_2017
```

```python
census1_pd_2017.drop('state', axis=1, inplace=True)
```

```python
census1_pd_2017.drop('county', axis=1, inplace=True)
```

```python
census1_pd_2017
```

```python
census1_pd_2017=census1_pd_2017[['NAME','2017 Pop','2017 Birth to Unmarried Women','High School Graduate(%)','Unemployment(%)',
                                 'Percent Under the Poverty lavel','Mean Income','Sex Ratio(per hundred female)','Mean Age']]
census1_pd_2017
```

```python
census1_pd_2017['NAME'] = census1_pd_2017['NAME'].str.rstrip('New Jersey')
census1_pd_2017
```

```python
census1_pd_2017['NAME'] = census1_pd_2017['NAME'].str.rstrip('County,')
```
