```python
import pandas as pd
```

```python
FBI_df = "../output_files/FBI.csv"
Agencies_df= "../output_files/agencies.csv"
```

```python
FBI_df = pd.read_csv(FBI_df, encoding="ISO-8859-1")
Agencies_df= pd.read_csv(Agencies_df, encoding="ISO-8859-1")
```

```python
FBI_df.head()
```

```python
Agencies_df.head()
```

```python
merged_df= pd.merge(FBI_df, Agencies_df, on="ori",how="left")
merged_df.head()
```

```python
crime_df= merged_df.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y', 'state_abbr'])
```

```python
#Changing county names from upper case to title case to reference with demography data
crime_df['county_name'] = crime_df['county_name'].str.title() 
crime_df.head()
```

```python
crime_df= crime_df.rename(columns={"ori": "ORI", "county_name":"county", "actual":"cases_reported", "cleared":"cases_cleared" })
crime_df.head()
```

```python
crime_df= crime_df[['county', 'agency_name', 'ORI','offense', 'cases_reported', 'cases_cleared', 'data_year' ]]
crime_df.head()
```

```python
crime_df.to_csv("../output_files/crime.csv", index=False, header=True)
```
